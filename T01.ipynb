### ¿Qué es una recesión? ###

Una **recesión** económica es una fase del ciclo económico en la que la actividad económica se reduce, en vez de aumentar. Generalmente se considera que existe recesión cuando la <u>tasa de variación del *PIB* es negativa durante dos trimestres consecutivos</u>

* Puntos clave
    * Las recesiones económicas son como frenazos en la economía, donde todo, desde el gasto y la producción se reduce.    * Las consecuencias finales son que algunas empresas quiebran y que los trabajadores pierden sus empleos.
    * Suele venir tras periodos de bonanza donde ha habido mucho gasto, inversión y endeudamiento.a.


Las recesiones existen desde hace miles de años. Entonces se conocían como periodos de **vacas flacas** por el gran peso que tenían la agricultura y ganadería. Se solían producir por sequías, enfermedades o guerras, e igualmente eran momentos muy duros porque decrecía la producción de alimentos y esto provocaba hambrunas.

Hoy en día las consecuencias no suelen ser tan dramáticas, pero igualmente la actividad económica cae. Lo que provoca que disminuya la proudcción y el consumo, la inversión y que por tanto aumente el desempleo.

### Principales causas de recesión ###

* Escacez de demanda
* Incertidumbre sobre el futuro
* Ciclos económicos
* Pérdida masiva de capitales

### Fórmula para calcular el PIB ###

Como anteriormente se menciona en la frase subrayada, la recesión se da cuando la tasa de variación del PIB es negativa en dos trimestres consecutivos.
Ahora veremos como se calcula:

$$ PIB = C + I + G + X – M $$

Donde las variables son:

* C = Consumo
* I = Inversión
* G = Gasto
* X = Exportaciones
* M = Importaciones
