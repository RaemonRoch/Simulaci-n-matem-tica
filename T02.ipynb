### 1. Vamos a graficar las siguientes funciones y determinar su dominio: ###

$$ f(x)=x^2-4x+2 $$
$$ f(x)=sin(3x) $$
$$ f(x)=e^{2x}-3 $$
$$ f(x)=\sqrt{x^2-4} $$
$$ f(x)=cos(x+\frac{\pi}{4}) $$

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-5,8)
y = x**2 - 3*x + 2

plt.plot(x,y)
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Función 1")
plt.fill_between(x, y.min() - 1, y.max() + 1, color='lightblue', alpha=0.3, label='Dominio de la función')
```

```python
def f(x):
    return np.sin(x)

x = np.linspace(-10,10,1000)
y = f(x)

plt.plot(x,y)
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Función 2")
```

```python
def f(x):
    return np.exp(2*x)-3
    
x = np.linspace(-3,3,100)
y = f(x)

plt.plot(x,y)
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Función 3")
```

```python
def f(x):
    return np.sqrt(x**2-4)

x = np.linspace(-10,10,1000)
y = f(x)

plt.plot(x,y)
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Función 4")
```

```python
def f(x):
    return np.cos(x + np.pi/4)

x = np.linspace(-10,10,1000)
y = f(x)

plt.plot(x,y)
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Función 5")
```

### 2. Vamos a graficar los siguientes grupos de funciones ###

$$ Grupo-1 $$
$$ y1 = 2x + 1 $$
$$ y2 = -x + 4 $$
$$ y3 = x^2 $$

$$ Grupo-2 $$
$$ y1 = sin(x) $$
$$ y2 = cos(x) $$

```python
x = np.linspace(-5,5)
y1 = 2*x + 1
y2 = -x + 4 
y3 = x**2

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.grid()
plt.plot(x,y1, "y", label="2x + 1")
plt.plot(x,y2, "g", label="-x + 4")
plt.plot(x,y3, "r", label="x^2")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Grupo 1")
plt.legend(loc="best")
```

```python
def f(x):
    return np.sin(x)
def g(x):
    return np.cos(x)
x = np.linspace(-10,10,1000)
y1 = f(x)
y2 = g(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.grid()
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Grupo 2")
plt.plot(x,y1, "r", label="sin(x)")
plt.plot(x,y2, "b", label="cos(x)")
plt.legend(loc="best")
```

### 3. Crea una función que cree una gráfica y dibuje 'k' líneas de forma f(x) = sin(ax) La función recibe: ###

* x1: valor inicial de 'x'
* x2: valor final de 'x'
* n: cantidad de puntos del vector 'x'
* a: valor inicial de image.png
* b: valor final de image.png
* k: cantidad de líneas

```python
def fun(x1,x2,n,a,b,k):
    x = np.linspace(x1,x2,n)
    av = np.linspace(a,b,k)
    for a in av:
        y = np.sin(a * x)
        plt.plot(x, y, label=f'sin({a:.2f}x)')
    plt.plot(x,y)
    plt.xlabel("Eje x")
    plt.ylabel("Eje y")
    plt.title("f(x)=sin(ax)")
    plt.legend()
fun(0, 2*np.pi, 1000, 1, 3, 3)
```
